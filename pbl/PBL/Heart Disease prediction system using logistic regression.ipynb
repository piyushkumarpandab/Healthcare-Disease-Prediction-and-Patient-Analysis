# Heart Disease prediction system using logistic regression

```python
#### Importing neccessary libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing   #### for machine learning
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Loading the dataset and handling the missing values
disease_df=pd.read_csv("framingham.csv")
disease_df.drop(columns=['education'], inplace = True, axis = 1)
disease_df.rename(columns = {'male':'Sex_male'}, inplace = True)
```

```python
disease_df.dropna(axis = 0, inplace = True)
disease_df
```

```python
### Printing the persons with diagnosed heart disease
print(disease_df.TenYearCHD.value_counts())
```

```python
### Splitting the data into test - train splits or sets
X = np.asarray(disease_df[['age','Sex_male','cigsPerDay',
                           'totChol','sysBP','glucose']])
y = np.asarray(disease_df['TenYearCHD'])
X = preprocessing.StandardScaler().fit(X).transform(X)
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.3,random_state=4)
print('Train set :', X_train.shape, y_train.shape)
print('Test set :', X_test.shape, y_test.shape)
```

```python
### Exploratory Data Analysis of Heart Disease Dataset (EDA)
### Ten Year's CHD Record of all the patients
plt.figure(figsize=(7,5))
sns.countplot(x='TenYearCHD', data=disease_df, hue='TenYearCHD', 
              palette='Set2', legend=False)
plt.show()
```

```python
### Counting number of patients affected by CHD where (0=not affected,1= affected)
laste = disease_df['TenYearCHD'].plot()
plt.show(laste)
```

```python
### blue bars = absence of heart disease
### white spaces = presence of heart disease
```

```python
### Fitting Logistic Regression Model For Prediction
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
### Evaluating Logistic Regression Model
from sklearn.metrics import accuracy_score
print(' Accuracy of the Model is = ',
      accuracy_score(y_test, y_pred))
```

```python
### Plotting Confusion Matrix to visualize the predictions 
from sklearn.metrics import confusion_matrix, classification_report
print('The Details for confusion matrix is = ')
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
conf_matrix = pd.DataFrame(data = cm,
                           columns = ['Predicted:0', 'Predicted:1'],
                           index = ['Actual:0', 'Actual:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix,annot=True,fmt='d',cmap="Blues")
plt.show()
```
